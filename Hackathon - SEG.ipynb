```python
# Filtering for manufacturing
manufacturing = df[df['NAICS Desc'] == "Manufacturing"]
manufacturing
```

```python
# Checking for nulls
print(manufacturing.isnull().sum())
```

```python
# Checking for nulls
print(manufacturing.isnull().sum())
```

```python
# Dropping rows with null values
manufacturing.dropna(inplace=True)
```

```python
print(manufacturing.isnull().sum())
```
